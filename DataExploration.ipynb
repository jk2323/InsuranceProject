```python
df.groupby(['smoker'])['charges'].mean()
```

```python
model_smoker = ols('charges ~ smoker', data=df).fit()
result_smoker = sm.stats.anova_lm(model_smoker, type=2)
```

```python
result_smoker
```

```python
model_sex = ols('charges ~ sex', data=df).fit()
result_sex = sm.stats.anova_lm(model_sex, type=2)
```

```python
result_sex
```

```python
model_region = ols('charges ~ region', data=df).fit()
result_region = sm.stats.anova_lm(model_region, type=2)
```

```python
result_region
```

```python
model_region_smoker = ols('charges ~ region + smoker + region:smoker', data=df).fit()
result_region_smoker = sm.stats.anova_lm(model_region_smoker, type=2)
result_region_smoker
```

```python
model_sex_smoker = ols('charges ~ sex + smoker + sex:smoker', data=df).fit()
result_sex_smoker = sm.stats.anova_lm(model_sex_smoker, type=2)
result_sex_smoker
```

```python
model_sex_region = ols('charges ~ sex + region + sex:region', data=df).fit()
result_sex_region = sm.stats.anova_lm(model_sex_region, type=2)
result_sex_region
```

```python
model_all = ols('charges ~ sex + smoker + region + smoker:sex + smoker:region + sex:region + sex:smoker:region', data=df).fit()
result_all = sm.stats.anova_lm(model_all, type=2)
result_all
```

```python
model_all.compare_lm_test(model_sex_smoker)
```

```python
model_all.summary()
```
